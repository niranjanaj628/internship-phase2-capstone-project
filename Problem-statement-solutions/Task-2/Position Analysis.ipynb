```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("cleaned_sports_dataset.csv")
```

```python
position_counts = df['Position'].value_counts()
max_position = position_counts.idxmax()
max_players = position_counts.max()
min_position = position_counts.idxmin()
min_players = position_counts.min()
total_players = position_counts.sum()
print(f"Total number of players: {total_players}")
print(f"Highest number of players: {max_players} players in position {max_position}")
print(f"Lowest number of players: {min_players} players in position {min_position}")

plt.figure(figsize=(16, 10))
#Bar plot
plt.subplot(1, 2, 1)
colors = sns.color_palette('pastel')
sns.barplot(x=position_counts.values, y=position_counts.index, palette=colors)
plt.title('Number of Players by Position')
plt.xlabel('Count of Players')
plt.ylabel('Position')

# Pie chart
plt.subplot(1, 2, 2)
position_counts.plot(kind='pie', autopct='%1.1f%%', startangle=100,colors=colors)
plt.title('Percentage of Players in Each Position')
plt.show()
```

```python
import pandas as pd
from scipy.stats import chisquare
position_counts = df['Position'].value_counts() #In each position
total_players = position_counts.sum()
expected_counts = total_players / len(position_counts)
print("\nTotal number of players:", total_players)
print("Expected count for uniform distribution per position:", expected_counts)
chi2_stat, p_value = chisquare(position_counts, f_exp=[expected_counts] * len(position_counts))
print("\nChi-square statistic:", chi2_stat) #formula obtained value
print("p-value:", p_value) #probability value (under null hypothesis : fits the expected distribution)
alpha = 0.05 #5%
if p_value < alpha:
    print("The distribution of players across positions is significantly different from a uniform distribution.")
else:
    print("The distribution of players across positions is not significantly different from a uniform distribution.")
```
